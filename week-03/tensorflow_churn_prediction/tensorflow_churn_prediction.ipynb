# Customer Churn Prediction System

## Project Objective
This project aims to build a simple neural network using TensorFlow to predict the probability of customer churn.

## What Does TensorFlow Solve?
- It builds models that learn from data instead of relying on manually written rules.
- It generates predictions by evaluating complex relationships between variables such as age, spending, subscription duration, and support requests.

## Topics Covered
- TensorFlow and the concept of “tensors”
- Dense layers (Sequential model)
- Activation functions (ReLU, Sigmoid)
- Train / Validation / Test split
- The difference between classification (0/1 churn) and regression

```python
# 2. data creation
import numpy as np

np.random.seed(42)

n = 2000

# Features
age = np.random.randint(18, 60, size=n)
monthly_expense = np.random.uniform(10, 500, size=n)
subs_months = np.random.randint(1, 61, size=n)
support_tickets = np.random.poisson(lam=2, size=n)

score = (
    0.03 * support_tickets +
    0.002 * monthly_expense -
    0.04 * subs_months +
    0.005 * (age - 40)
)

prob = 1 / (1 + np.exp(-score))

y = (prob > 0.5).astype(int)

X = np.column_stack([age, monthly_expense, subs_months, support_tickets]).astype(np.float32)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("Churn rate:", y.mean())
```

```python
# 3.train/test
from sklearn.model_selection import train_test_split

X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, random_state=42
)

# Control
print("Train size:", X_train.shape)
print("Validation size:", X_val.shape)
print("Test size:", X_test.shape)
```

```python
# 4. TensorFlow Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(16, activation="relu", input_shape=(X_train.shape[1],)),
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")
])

model.summary()
```

```python
# 5. Activation functions
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([

    # Hidden layer 1
    Dense(16, activation="relu", input_shape=(X_train.shape[1],)),

    # Reason for using ReLU:
    # - it enables faster learning
    # - it is the most widely used activation function in deep learning

    # Hidden layer 2
    Dense(8, activation="relu"),

    # Output layer
    Dense(1, activation="sigmoid")

    # Reason for using Sigmoid:
    # - it compresses the output into the range between 0 and 1
    # - since churn prediction is a binary classification problem,
    #   the model produces a "probability of churn"
])

model.summary()
```

```python
# 6. Model Compiling
model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=["accuracy"]
)
```

```python
# 7. Model training
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=30,
    batch_size=16,
    verbose=1
)
```

```python
# 8. Model evaluation
test_loss, test_accuracy = model.evaluate(X_test, y_test)

print("Test Loss:", test_loss)
print("Test Accuracy:", test_accuracy)
```

```python
# 9. New customer
new_customer = np.array([[35, 250, 12, 3]], dtype=np.float32)

prediction = model.predict(new_customer)

print("Churn rate:", prediction[0][0])
```
