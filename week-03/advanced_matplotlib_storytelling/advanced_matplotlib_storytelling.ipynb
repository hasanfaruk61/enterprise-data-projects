# Advanced Data Visualization & Storytelling Dashboard

## Project Objective
The goal of this project is to combine different business insights into a single figure to create a dashboard that allows management to understand the overall situation at a glance.

## Why Advanced Visualization is Needed
- A single chart usually answers only one question, while management often needs to see trends, comparisons, and distributions at the same time.
- A multi-chart dashboard approach helps identify cause-and-effect relationships much faster.
- Annotations and textual guidance help direct the viewer’s attention to critical insights.

## Matplotlib Concepts Used
- Subplots (advanced usage)
- Visual enhancements
- Advanced visual design
- Different chart types# Advanced Data Visualization & Storytelling Dashboard

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

# 1) Monthly sales
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
monthly_sales = np.random.randint(80000, 200000, size=12)

# 2) Revenue by product
products = ["Smartphone", "Laptop", "Smart TV", "Tablet", "Smartwatch"]
product_revenue = np.random.randint(40000, 160000, size=5)

# 3) Performance by region
regions = ["Northeast", "Midwest", "South", "West"]
region_sales = np.random.randint(150000, 500000, size=4)

# 4) Distribution of order amounts
orders_count = np.random.randint(1, 25, size=60)
avg_basket = np.random.randint(20, 250, size=60)
order_amounts = np.random.randint(10, 300, size=300)

# Create DataFrames
sales_df = pd.DataFrame({"Month": months, "Sales": monthly_sales})
product_df = pd.DataFrame({"Product": products, "Revenue": product_revenue})
region_df = pd.DataFrame({"Region": regions, "Sales": region_sales})
customer_df = pd.DataFrame({"OrdersCount": orders_count, "AvgBasket": avg_basket})

sales_df.head()
```

```python
# 3 Subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Business Dashboard - Sales & Customer Insights", fontsize=14)

# 1) Line Plot - Monthly sales trend (top left)
axes[0, 0].plot(sales_df["Month"], sales_df["Sales"], marker="o")
axes[0, 0].set_title("Monthly Sales Trend")
axes[0, 0].set_xlabel("Month")
axes[0, 0].set_ylabel("Sales")
axes[0, 0].grid(True)

# 2) Bar Plot - Revenue by product (top right)
axes[0, 1].bar(product_df["Product"], product_df["Revenue"])
axes[0, 1].set_title("Revenue by Product")
axes[0, 1].set_xlabel("Product")
axes[0, 1].set_ylabel("Revenue")
axes[0, 1].tick_params(axis="x", rotation=20)

# 3) Pie Chart - Sales by region (bottom left)
axes[1, 0].pie(
    region_df["Sales"],
    labels=region_df["Region"],
    autopct="%1.1f%%",
    startangle=90
)

axes[1, 0].set_title("Sales Share by Region")
axes[1, 0].axis("equal")

# 4) Scatter Plot - Customer behavior (bottom right)
axes[1, 1].scatter(customer_df["OrdersCount"], customer_df["AvgBasket"])
axes[1, 1].set_title("Customer Behavior (Orders vs Avg Basket)")
axes[1, 1].set_xlabel("Orders Count")
axes[1, 1].set_ylabel("Average Basket")

fig.patch.set_facecolor("#D8D8D8")

plt.show()
```

### 1) Monthly Sales Trends (Top Left — Line Plot)

The dashboard begins with the **monthly sales trend**.
This chart shows how sales change throughout the year.

A line plot is used because it **best illustrates changes over time and highlights increases or decreases clearly.**

---

### 2) Revenue by Product

The second chart compares the total revenue generated by each product.
A bar chart is used because it **clearly shows comparisons between different categories.**

---

### 3) Sales Distribution by Region

The third chart shows how sales are distributed across different regions.
A pie chart is used because it **visually represents each region’s share of the total sales in a quick and intuitive way.**

---

### 4) Customer Behavior

The final chart analyzes customer behavior.
It shows the relationship between the number of orders and the average basket value.

A scatter plot is used because it **reveals relationships and potential patterns between two variables.**
