# Business Performance Visualization Dashboard

## Objective

The goal of this project is to transform raw business performance data into meaningful visual insights using Matplotlib.

The objective is not only to create charts, but also to clearly and accurately communicate trends, comparisons, and strategic signals.

## Why Is Data Visualization Important?

Without visualization, interpreting data can be difficult.

Well-designed charts:

- Reveal trends
- Support business decisions
- Improve communication with stakeholders
- Reduce cognitive load

## Tools Used

- Python
- NumPy
- Pandas
- Matplotlib (Line, Bar, Subplots)

```python
# 2. Creating Data sets
import pandas as pd
import numpy as np

# Months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Monthly Sale data set
monthly_sales = np.random.randint(200000, 900000, size=len(months))

# Product based sale revenue
products = ["Phone", "Laptop", "Tablet", "Watch", "Accessories"]
product_revenue = np.random.randint(80000, 450000, size=len(products))

# Region based revenue
regions = ["West", "Midwest", "South", "Northeast"]
region_sales = np.random.randint(150_000, 700_000, size=len(regions))

df_monthly = pd.DataFrame({"Month": months, "Sales": monthly_sales})
df_products = pd.DataFrame({"Product": products, "Revenue": product_revenue})
df_regions = pd.DataFrame({"Region": regions, "Sales": region_sales})

df_monthly, df_products, df_regions
```

```python
# 3. Line Plot
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))

plt.plot(df_monthly["Month"],
         df_monthly["Sales"],
         marker='o',
         linestyle='-',
         linewidth=2)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.grid(True)

plt.xticks(rotation=45)

plt.show()
```

```python
# 4.Bar Plot
plt.figure(figsize=(18,5))

plt.bar(df_products["Product"],
        df_products["Revenue"])

plt.title("Product Revenue Comparison")
plt.xlabel("Product")
plt.ylabel("Revenue ($)")

plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Graphic Customization
plt.figure(figsize=(9,5))

bars = plt.bar(
    df_products["Product"],
    df_products["Revenue"],
    color="skyblue",
    edgecolor="black"
)

plt.title("Product Revenue Comparison - 2024", fontsize=14, fontweight="bold")
plt.xlabel("Product Category", fontsize=12)
plt.ylabel("Revenue ($)", fontsize=12)

plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# 6. Subplots
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Region data (US regions) ---
regions = ["Northeast", "Midwest", "South", "West"]
region_sales = np.random.randint(30000, 120000, size=len(regions))

df_regions = pd.DataFrame({
    "Region": regions,
    "Sales": region_sales
})

# --- Dashboard ---
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1) Line plot: Monthly sales
axes[0, 0].plot(months, monthly_sales, marker="o", linestyle="-")
axes[0, 0].set_title("Monthly Sales Trend")
axes[0, 0].set_xlabel("Month")
axes[0, 0].set_ylabel("Sales ($)")
axes[0, 0].grid(True, linestyle="--", alpha=0.6)
axes[0, 0].tick_params(axis="x", rotation=45)

# 2) Bar plot: Product revenue
axes[0, 1].bar(df_products["Product"], df_products["Revenue"], edgecolor="black")
axes[0, 1].set_title("Revenue by Product")
axes[0, 1].set_xlabel("Product")
axes[0, 1].set_ylabel("Revenue ($)")
axes[0, 1].grid(axis="y", linestyle="--", alpha=0.6)
axes[0, 1].tick_params(axis="x", rotation=45)

# 3) Bar plot: Region sales
axes[1, 0].bar(df_regions["Region"], df_regions["Sales"], edgecolor="black")
axes[1, 0].set_title("Sales by US Region")
axes[1, 0].set_xlabel("Region")
axes[1, 0].set_ylabel("Sales ($)")
axes[1, 0].grid(axis="y", linestyle="--", alpha=0.6)
axes[1, 0].tick_params(axis="x", rotation=20)

# 4) Summary "visual": show the best month and best product as text
best_month_idx = int(np.argmax(monthly_sales))
best_month = months[best_month_idx]
best_month_value = monthly_sales[best_month_idx]

best_product_idx = int(df_products["Revenue"].idxmax())
best_product = df_products.loc[best_product_idx, "Product"]
best_product_value = df_products.loc[best_product_idx, "Revenue"]

axes[1, 1].axis("off")
axes[1, 1].text(
    0.05, 0.75,
    f"Best Month: {best_month} (${best_month_value:,.0f})",
    fontsize=12
)
axes[1, 1].text(
    0.05, 0.55,
    f"Top Product: {best_product} (${best_product_value:,.0f})",
    fontsize=12
)
axes[1, 1].text(
    0.05, 0.35,
    f"Total Annual Sales: ${sum(monthly_sales):,.0f}",
    fontsize=12
)

plt.suptitle("Business Performance Dashboard", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.show()
```
