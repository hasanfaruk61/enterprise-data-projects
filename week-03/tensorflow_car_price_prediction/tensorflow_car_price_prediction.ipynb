# Car Price Prediction System (TensorFlow)

## Project Objective

The purpose of this project is to develop a machine learning model that can estimate the price of a car by using its features.

The model learns to predict car prices by analyzing several key vehicle characteristics. These features include:

- Model year
- Mileage
- Engine size
- Horsepower
- Vehicle age

## Why Is This a Regression Problem?

In this project, the predicted value is the **car price**, and prices are continuous numerical values.

For this reason, the problem is defined as a **regression problem**.

In regression problems, the model predicts a numerical value rather than a category.

## TensorFlow Topics Used

This project covers the following TensorFlow topics:

- Data preparation
- Train / test split
- Building an artificial neural network model
- Model training
- Model evaluation
- Making predictions for new data

```python
# 2.Data Creation
import numpy as np
import pandas as pd

np.random.seed(40)
n = 1000

year = np.random.randint(2005, 2024, n)
km = np.random.randint(5000, 200000, n)
engine = np.random.randint(1.0, 4.0, n)
hp = np.random.randint(70, 400, n)
age = 2024 - year

price = (
    30000
    + (year - 2005) * 800
    - km * 0.05
    + engine * 5000
    + hp * 120
    + np.random.normal(10, 5000, n)
)

data =  pd.DataFrame({
    "year": year,
    "km": km,
    "engine": engine,
    "hp": hp,
    "age": age,
    "price": price
})

X = data.drop("price", axis=1)
y = data["price"]
```

```python
# 3.Train/Test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# 4.Creating a model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(32, activation="relu"),
    Dense(16, activation="relu"),
    Dense(1)
])

model.summary()
```

```python
# 5.Compile
model.compile(
    loss="mean_squared_error",
    optimizer="adam",
    metrics=["mae"]
)
```

```python
# 6.Training
history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_split=0.2,
    verbose=1
)
```

```python
# 7.Model Evaluation
test_loss, test_mae = model.evaluate(X_test, y_test)

print("Test Loss (MSE):", test_loss)
print("Test MAE:", test_mae)

# Since the data is synthetic, it usually does not cause major issues
```

```python
# 8.Model Predictions
new_car = pd.DataFrame({
    "year": [2020],
    "km": [45000],
    "engine": [2.0],
    "hp": [150],
    "age": [2024 - 2020],
})
new_car = (new_car - X.mean()) / X.std()
predicted_price = model.predict(new_car)
print("Predicted car price:", predicted_price[0][0])
```

## Car Price Analysis

Some of the most important factors affecting car prices are the model year, mileage, and engine power.

In general, newer vehicles tend to have higher prices. Additionally, cars with lower mileage are usually considered more valuable in the second-hand market.

Engine size and horsepower can also influence the price, since more powerful engines often imply higher performance and higher costs.

The model tends to predict mid-range priced vehicles more accurately. However, prediction errors may increase for very expensive or very cheap cars, since the model may not have learned these extreme values sufficiently.
