Bu proje; NumPy array mantığını, metodları ve dizi işlemlerini kullanarak **gerçekçi bir satış verisi analizi** yapmayı hedefler.

#### Numpy Neden Kullanilir?
NumPy, Python’da özellikle sayısal işlemleri hızlı ve verimli yapmak için kullanılır.

* Büyük sayı dizileriyle (array) çalışmayı sağlar
* Matematiksel işlemleri çok hızlı yapar
* Bilimsel hesaplamaların temelini oluşturur

#### Kullanilan Numpy Konulari
* Array oluşturma
* Array özellikleri (shape, ndim, dtype)
* Indexing & slicing
* sum, mean, max, min
* axis kullanımı
* reshape, transpose, flatten

```python
# Importing Numpy and creating data
import numpy as np

# random sales matrix(7*5)
sales = np.random.randint(50, 200, size=(7, 5))
sales
```

```python
# Arrays
shape_of_sales = sales.shape
dim_of_sales = sales.ndim
type_of_sales = sales.dtype
size_of_sales = sales.size

print(f"shapeof sales: {shape_of_sales}")
print(f"dimension of sales: {dim_of_sales}")
print(f"type of sales: {type_of_sales}")
print(f"size of sales: {size_of_sales}")

```

```python
#Indexing and slicing
first_day_of_sales = sales[0]
second_product_weekly_sales = sales[:,2]
first_three_day_of_sales = sales[:3]

print(f"Sales for 1th day: {first_day_of_sales}")
print(f"Weekly sales for 2nd product: {second_product_weekly_sales}")
print(f"Sales for first 3 day: {first_three_day_of_sales}")
```

```python
# Basic Numpy Methods
total_sales = sales.sum()
avg_sale = sales.mean()
max_sale = sales.max()
min_sale = sales.min()

print(f"Total sales: {total_sales}")
print(f"Average sale: {avg_sale}")
print(f"Max sale: {max_sale}")
print(f"Min sale: {min_sale}")
```

```python
# Using Axis
total_for_weekly_product_sales = sales.sum(axis=0)
total_for_daily_product_sales = sales.sum(axis=1)

# axis = 0 -> sütunlar boyunca islem yapar
# axis = 1 -> satirlar boyunca islem yapar
```

```python
# Numpy array methods
reshaped_sales = sales.reshape(5,7)
transposed_sales = sales.T
flatted_sales = sales.flatten()
# ravel_sales = sales.ravel()

print(f"Reshaped sales: {reshaped_sales}")
print(f"Transposed sales: {transposed_sales}")
```

```python
# Outputs
best_sale_product_index = total_for_daily_product_sales.argmax() + 1
best_sale_day_index = total_for_weekly_product_sales.argmax() + 1


print(f"Weekly Total Sale: {total_sales}")
print(f"Best Sale Product: product {best_sale_product_index}")
print(f"Best Sale Day: day {best_sale_day_index}")
```
