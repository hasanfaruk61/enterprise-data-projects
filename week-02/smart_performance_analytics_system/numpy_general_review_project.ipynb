## Projenin Amaci
Bir şirketin 8 çalışana ait 6 aylık performans skorları NumPy ile analiz edilecektir.
## NumPy Neden Kullanilir?
NumPy'in kullanim amaclari:

* Hızlı sayısal hesaplama yapmak için
* Büyük veri dizileriyle (array/matris) çalışmak için
* Matematiksel işlemleri döngüsüz ve performanslı yapmak için
* Lineer cebir ve bilimsel hesaplamalar için
* Veri bilimi ve makine öğrenmesinin temelini oluşturmak için

## Kapsanan NumPy Konulari
Bu projede **yalnızca** aşağıdaki NumPy konuları kullanılacaktır:

- NumPy giriş
- NumPy array oluşturma
- NumPy metodları
- NumPy dizi (array) metodları
- NumPy indeksleme
- Matrix indeksleme
- NumPy operasyonlar (vektörel işlemler)


```python
# 2. Importing numPy and creating a random performance matrix(8x6)
import numpy as np
from numpy.ma.core import argmax

performance_matrix = np.random.randint(0,100, size=(8,6))
print(performance_matrix)
```

```python
# 3. shape, ndim and dtype
shape_of_matrix = performance_matrix.shape
ndim_of_matrix = performance_matrix.ndim
type_of_matrix = performance_matrix.dtype

print(shape_of_matrix)
print(ndim_of_matrix)
print(type_of_matrix)
```

```python
# 4. Indexing in NumPy
first_employee_all_months = performance_matrix[0]
last_month_all_employees = performance_matrix[:,-1]
first_three_employees_first_two_months = performance_matrix[:3, :2]

print(first_employee_all_months)
print(last_month_all_employees)
print(first_three_employees_first_two_months)
```

```python
# 5. Indexing matrix
employee_index = 3
month_index = 5

performance_score = performance_matrix[employee_index, month_index]

sub_matrix = performance_matrix[:3, :4]

print(f"5th month score of 3rd Employee: {performance_score}")
print(sub_matrix)
```

```python
# 6. Adding 5 points to all scores
adding_5_points_to_all_scores = performance_matrix + 5
normalizing_all_scores = performance_matrix / 100

# Comparing 3rd and 5th month scores
month4 = performance_matrix[:, 3]
month5 = performance_matrix[:, 4]

comp = (month5 > month4)

print(adding_5_points_to_all_scores)
print(normalizing_all_scores)
print(comp)
```

```python
# 7. NumPy Methods
avg_score_of_employees = performance_matrix.mean(axis=1)
avg_score_of_months = performance_matrix.mean(axis=0)
max_performance = performance_matrix.max()
min_performance = performance_matrix.min()

# axis=1 → satırlar boyunca işlem yapar. Her satirin yani calisanin ortalamasini alir
# axis=0 → sütunlar boyunca işlem yapar. Her sütunun yani ayin ortalamasini alir.

print(f"Average Scores of Employee: {avg_score_of_employees}")
print(f"Average Scores of Months: {avg_score_of_months}")
print(f"Max Scores of Employee: {max_performance}")
print(f"Min Scores of Employee: {min_performance}")
```

```python
# 8. NumPy series methods
reshaped_matrix = performance_matrix.reshape(4,12)
transposed_matrix = performance_matrix.transpose()
flattened_matrix = performance_matrix.flatten()

print(reshaped_matrix)
print(transposed_matrix)
print(flattened_matrix)
```

```python
# General Performance Summary
best_employee_index = avg_score_of_employees.argmax()
highest_month_average_index = avg_score_of_months.argmax()
general_average = performance_matrix.mean()

print(f"Best Employee: Employee{best_employee_index + 1}")
print(f"Highest Month: Month{highest_month_average_index + 1}")
print(f"General Average Performance: {general_average}")
```
