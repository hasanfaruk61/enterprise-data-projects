```python
# 2.Importing pandas and creating series for monthly expenses
import pandas as pd
import numpy as np

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
spending = np.random.randint(12500, 800000, len(months))

monthly_spending = pd.Series(spending, index=months)

print(f"Months: {months}")
print(f"\nExpenses: {spending}")
print(f"\nMonthly Expenses: \n{monthly_spending}")
```

```python
# 3.Characteristics of Series
values_of_monthly_spending = monthly_spending.values
indexes_of_monthly_spending = monthly_spending.index
type_of_monthly_spending_series = monthly_spending.dtype
avg_monthly_spending = round(monthly_spending.mean(), 2)

print(f"Values of monthly spending: {values_of_monthly_spending}")
print(f"Index Of Monthly Expenses: {indexes_of_monthly_spending}")
print(f"Type of Monthly spending: {type_of_monthly_spending_series}")
print(f"Average Monthly spending: {avg_monthly_spending}")
```

```python
# 4.Creating a dataframe
customer_names = ["Joe","Johny","Brad","Chris","Jennifer","Sam","George","Natalie","Sally","Michael","Christina","Marry"]
cities = ["New York","Los Angeles","Chicago","Houston","Phoenix","Philadelphia","San Antonio","San Diego","Dallas","San Jose","Austin","Jacksonville"]
monthlySpendings = np.random.randint(12500, 800000, len(customer_names))
membershipTypes = ["Basic","Standard","Premium","Gold","Silver","Platinum","VIP","Student","Family","Corporate","Trial","Lifetime"]

data = {
    "CustomerName": customer_names,
    "City": cities,
    "MonthlySpending" : monthlySpendings,
    "MembershipType": membershipTypes,
}
df = pd.DataFrame(data)
df
```

```python
# 5.Indexing DataFrame
new_index = df.set_index("CustomerName")

print(f"New Index: \n{new_index}")
print(f"\nChoosing a row via index: \n{new_index.loc['Joe']}")
```

```python
# 6.Changing Index
index_reset = new_index.reset_index()

setting_new_index = index_reset.set_index("City")

print(f"Index was reset: \n{index_reset}")
print(f"\nSetting new index: \n{setting_new_index}")
```

```python
# 7.Creating Multi Index
df_multi_index = df.set_index(["City","MembershipType"])
df_multi_index
```

```python
# 8.General Summary
number_of_customers = df.shape[0]
highest_monthly_spending = df["MonthlySpending"].max()

print(f"Number of customers: {number_of_customers}")
print(f"Highest monthly spending: {highest_monthly_spending}")
```
