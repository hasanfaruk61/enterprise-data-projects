```python
# 2. Book Class
class Book():
    def __init__(self, title, author, year, is_available=True ):
        self.title = title
        self.author = author
        self.year = year
        self.is_available = is_available

    def get_info(self):
        print(f"Kitap bilgisi: {self.title}, {self.author}, {self.year}, {self.is_available}")
```

```python
# 3. Member class
class Member():
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id

    def borrow_book(self, book):
        if book.is_available:
            book.is_available = False
            print("Kitabi alabilirsiniz")
        else:
            print("Kitap mevcut degil!")

    def return_book(self, book):
         if not book.is_available:
             book.is_available= True
             print("Kitabi geri getirildi")
         else:
             print("Kitap mevcut!")
```

```python
# 4. Default Values
# Default deger kullanimi zorunlu olmayan alanlari opsiyonel yapmamiza yardim eder.
# Nesne olusturmayi kolaylastirir.
book1 = Book("Cali Kusu", "Resat Nuri Güntekin", 1937)
book1.get_info()
```

```python
# 5. Digital Book class
class DigitalBook(Book):
    def __init__(self, title, author, year, file_size, is_available=True):
        super().__init__(title, author, year, is_available)
        self.file_size = file_size

    def download(self):
        print("Dosya indiriliyor...")
```

```python
# 6. Test
book1 = Book("Kürk Mantolu Madonna", "Sebahattin Ali", 1943)
book2 = Book("Suç ve Ceza", "Fyodor Dostoyevski", 1866)
digital_book1 = DigitalBook("Fareler ve İnsanlar", "John Steinbeck", 1937, file_size=15)

member1 = Member("Ahmet", 1)

book1.get_info()
book2.get_info()
digital_book1.get_info()

member1.borrow_book(book1)
member1.borrow_book(book1)

member1.return_book(book1)

digital_book1.download()

```
