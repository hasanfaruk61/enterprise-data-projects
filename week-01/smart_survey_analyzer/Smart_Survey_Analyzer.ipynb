```python
import random

scores = []
while len(scores) < 20:
    score = random.randint(0, 6)

    if score == 0:
        break

    if score == 6:
        continue
    scores.append(score)
```

```python
# 2. Data Check
total_valid_score = len(scores)

# While döngüsü kullanildi cünkü
# Kac denemde 20 gecerli score toplanacagini önceden bilmiyoruz
# 0 ve 6 kosulunu kontrol edebilip gerektiginde döngüyü kirmak ya da devam ettirmek istiyoruz
```

```python
# 3. Average Score
if len(scores) > 0:
    total_score = sum(scores)
    avg_score = total_score / len(scores)
    rounded_avg_score = round(avg_score, 2)
else:
    print("No valid score found")
```

```python
# 4. Min/Max
if len(scores) > 0:
    max_score = max(scores)
    min_score = min(scores)
else:
    print("No valid score found")
```

```python
# 5. Satisfaction Analysis
memnun = 0
kararsiz = 0
memnun_degil = 0

if len(scores) > 0:
    for score in scores:
        if score >= 4:
            memnun += 1
        elif score == 3:
            kararsiz += 1
        else:
            memnun_degil += 1
else:
    print("No valid score found")
```

```python
# 6: Early Warning
score_check = 0

for score in scores:
    if score == 1:
        score_check += 1
    else:
        score_check = 0
    if score_check == 3:
        print("⚠️ Kritik memnuniyetsizlik tespit edildi")
        break
```

```python
print(f"Toplam gecerli cevap: {total_valid_score}")
print(f"Ortalama puan: {rounded_avg_score}")
print(f"Min puan: {min_score}")
print(f"Max puan: {max_score}")
print("")
print(f"Memnun: {memnun}")
print(f"Kararsiz: {kararsiz}")
print(f"Memnun_degil: {memnun_degil}")
```
