# Polymorphism nedir, neden gereklidir?

Polymorphism (çok biçimlilik), nesne yönelimli programlamada aynı isimdeki bir metodun farklı sınıflarda farklı davranışlar sergileyebilmesidir. Bu sayede
ortak bir arayüz korunurken, her nesne kendi özel davranışını uygulayabilir.

Bu projede send() metodunun Email, SMS ve Push sınıflarında farklı çıktılar üretmesi, polymorphism’in doğrudan bir örneğidir. Aynı metod çağrısı yapılmasına rağmen her nesne kendi davranışını sergilemektedir.

# Kullanilan Python konulari

Bu projede **yalnızca** aşağıdaki başlıklar kullanılacaktır:

- Polymorphism
- Özel metodlar (`__str__`, `__len__`)
- Hataları ele alma (`try / except`)
- Standart kütüphaneler (`datetime`, `random`)
- Modüller ve paketler (dosya bazlı ayırma mantığı)


```python
from notifications import EmailNotification, SMSNotification, PushNotification
from utils import generate_message
```

```python
message = generate_message()
notifications = [
    EmailNotification(message),
    SMSNotification(message),
    PushNotification(message)
]
```

```python
for notification in notifications:
    notification.send()
```
