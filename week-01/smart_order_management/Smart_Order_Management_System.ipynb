```python
# 1.Raw order analysis
orders = [
    {
"customer":"Joe",
"items": ["Laptop","Mouse","Mouse"],
"total":18500,
"is_paid":True
    },
    {
"customer":"Sally",
"items": ["Keyboard"],
"total":2500,
"is_paid":False
    },
    {
"customer":"Kim",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    },
    {
"customer":"Reynold",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    }
]
```

```python
# 2.Paid & unpaid order count
order_count = len(orders)

paid_order_count = 0
unpaid_order_count = 0

for order in orders:
    if order['is_paid']:
        paid_order_count += 1
    else:
        unpaid_order_count += 1
```

```python
# 3.Customer list
customer_list = []

for order in orders:
    if order['customer']:
        customer_list.append(order['customer'])
distinct_customers = list(set(customer_list))
```

```python
# 4.Product Analysis
all_product_list = []

for order in orders:
    for item in order['items']:
        all_product_list.append(item)

duplicates = list(set(item for item in all_product_list if all_product_list.count(item) > 1))
unique_product_list = set(all_product_list)
```

```python
# 5.Order Status Check
confirmed_orders = []
unconfirmed_orders = []

for order in orders:
    if order['is_paid']:
        print(order['customer'], "- Order confirmed")
    else:
        print(order['customer'], "- Order not confirmed")
```

```python
# 6.Secured Order Summary
order_sum = []
for order in orders:
    new_tuple = (order['customer'], order['total'])
    order_sum.append(new_tuple)

# Tuples are designed to preserve data integrity.
# Once created, their content cannot be modified.
# This prevents accidental data changes.
```

```python
print("Total orders: " + str(paid_order_count + unpaid_order_count))
print("Paid orders: " + str(paid_order_count))
print("Unpaid orders: " + str(unpaid_order_count))

print("Unique products: " + str(unique_product_list))
print("Order summaries: " + str(order_sum))
```
