# Proje Tanimi

## Projenin Amaci
Kullanıcıdan alınan harcama tutarlarını güvenli bir şekilde toplayıp; toplam, ortalama, en yüksek ve en düşük harcama gibi temel bütçe özetlerini üretmek.

## Gercek hayatta ne ise yarar
- Kişisel bütçe takibi (günlük/haftalık harcamalar)
- Harcama alışkanlığı analizi (ortalama, uç değerler)
- Basit finansal raporlama (toplam, min/max)

## Kullanilan Python Konulari
- Fonksiyonlar (kod tekrarını azaltma, modüler yapı)
- input/print (kullanıcı etkileşimi)
- try/except (hatalı girişleri yönetme)
- Liste kullanımı (harcamaları saklama)
- sum, len, round, max, min (temel istatistiksel özetler)
- *args (değişken sayıda argüman ile toplama)
- Scope (global/local değişken davranışı)

```python
# 2. Get Expenses Function
def get_expenses():
    expenses = []

    while True:
        try:
            count = int(input("Kaç harcama gireceksiniz? "))
            if count <= 0:
                print("Lütfen 1 veya daha büyük bir sayı girin.")
                continue
            break
        except ValueError:
            print("Geçersiz giriş. Lütfen tam sayı girin.")

    i = 0
    while i < count:
        try:
            value = float(input(f"{i+1}. harcama tutarı: "))
            if value < 0:
                print("Harcama negatif olamaz. Tekrar deneyin.")
                continue
            expenses.append(value)
            i += 1
        except ValueError:
            print("Geçersiz giriş. Lütfen sayı giriniz!")
    return expenses
```

```python
# 3. Summary and Average
def calculate_summary(expenses):

    if len(expenses) == 0:
        return 0, 0
    total = sum(expenses)
    avg = round(total / len(expenses), 2)

    return total, avg
```

```python
# 4. Flexible Summary with *args
def flexible_total(*args):
    total_of_expenses = sum(args)
    count_of_expenses = len(args)
    return f"Girilen Harcamalarin Toplami: {total_of_expenses}, Girilen Harcama Sayisi: {count_of_expenses}"
```

```python
# 5. Max/Min
def find_extremes(expenses):
   if len(expenses) == 0:
       return 0, 0
   max_expense = max(expenses)
   min_expense = min(expenses)
   return max_expense, min_expense
```

```python
# 6.Global Variable
global_currency = "TL"

def change_currency():
    global_currency = "EUR"
    print("Local currency: ", global_currency)

print(global_currency)

# Global değişkenler fonksiyon dışında tanımlanır ve programın her yerinden okunabilir.
# Fonksiyon içinde bir değişkene atama yapılırsa, Python onu varsayılan olarak local (yerel) kabul eder.
# Bu nedenle fonksiyon içinde yapılan atama global değişkeni değiştirmez.
```

```python
# Main Flow
expenses = get_expenses()
total, avg = calculate_summary(expenses)
max_expense, min_expense = find_extremes(expenses)

print(f"Toplam harcama: {total} {global_currency}")
print(f"Ortalama harcama: {avg} {global_currency}")
print(f"En yüksek harcama: {max_expense} {global_currency}")
print(f"En dusuk harcama: {min_expense} {global_currency}")

```
