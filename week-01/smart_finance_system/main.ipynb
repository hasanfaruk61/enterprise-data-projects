```python
from unicodedata import category

from models import Budget
from utils import get_user_expenses

# sabit limit belirlendi
LIMIT = 10000
budget = Budget(LIMIT)

print(f"Bütce limiti: {LIMIT} TL\n")

expenses = get_user_expenses()

for expense in expenses:
    budget.add_expense(expense)
```
