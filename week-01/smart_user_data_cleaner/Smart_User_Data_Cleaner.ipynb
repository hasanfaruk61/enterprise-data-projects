```python
raw_data ="  SalLY  rOgeRS ;  23 ;  1.78 ;  sallyROGERS@GMAIL.Com   "
```

```python
# 1. General Cleaning
cleaned_data = raw_data.strip()

sliced_data = cleaned_data.split(";")
```

```python
# 2. Firstname & Lastname Correction
raw_name = sliced_data[0].lower()
raw_first_name = raw_name.split()[0]
raw_last_name = raw_name.split()[1]

cleaned_first_name = raw_first_name.strip()
cleaned_last_name = raw_last_name.strip()

capitalized_firstname = cleaned_first_name.capitalize()
capitalized_lastname = cleaned_last_name.capitalize()

user_name = capitalized_firstname + " " + capitalized_lastname
```

```python
# 3. Age processing
raw_age = sliced_data[1].strip()
int_raw_age = int(raw_age)
ten_year_older = int_raw_age + 10
user_age = float(ten_year_older)
```

```python
# 4. Height analysis
raw_height = sliced_data[2].strip()

float_raw_height = float(raw_height)
height_in_cm = float_raw_height * 100
```

```python
# 5- Email cleaning
raw_email = sliced_data[3].strip().lower()

q_index = raw_email.find("@")
user_name = raw_email[:q_index]
user_email_code = user_name[:3]
```

```python
print(f"User: {user_name}")
print(f"Age (in 10 years): {user_age}")
print(f"Height: {height_in_cm}")
print(f"Email user code: {user_email_code}")
```
